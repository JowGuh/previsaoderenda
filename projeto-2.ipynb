#### Entendimento dos dados - Univariada
Nesta etapa tipicamente avaliamos a distribuição de todas as variáveis. 

```python
def analise_univariada(df):
    for coluna in df.columns:
        if df[coluna].dtype == 'object' or df[coluna].nunique() < 20:
            
            plt.figure(figsize=(10, 6))
            sns.countplot(data=df, x=coluna, palette='Set2')
            plt.title(f'Distribuição da variável {coluna}')
            plt.xticks(rotation=45)
            plt.show()
        else:
            
            plt.figure(figsize=(10, 6))
            sns.histplot(df[coluna], kde=True, color='blue')
            plt.title(f'Distribuição da variável {coluna}')
            plt.show()


analise_univariada(renda)
```

### Entendimento dos dados - Bivariadas




```python
def analise_bivariada(df, var1, var2):
    
    
    if pd.api.types.is_numeric_dtype(df[var1]) and pd.api.types.is_numeric_dtype(df[var2]):
        plt.figure(figsize=(10, 6))
        sns.scatterplot(data=df, x=var1, y=var2, color='green')
        plt.title(f'Relação entre {var1} e {var2}')
        plt.show()
    
    
    elif df[var1].dtype == 'object' and pd.api.types.is_numeric_dtype(df[var2]):
        plt.figure(figsize=(10, 6))
        sns.barplot(data=df, x=var1, y=var2, palette='Set2')
        plt.title(f'{var1} vs {var2}')
        plt.xticks(rotation=45)
        plt.show()
    
    
    elif df[var1].dtype == 'object' and df[var2].dtype == 'object':
        plt.figure(figsize=(10, 6))
        sns.countplot(data=df, x=var1, hue=var2, palette='Set2')
        plt.title(f'Distribuição de {var2} por {var1}')
        plt.xticks(rotation=45)
        plt.show()
    else:
        print(f"Não há suporte para analisar a combinação de {var1} e {var2}")


analise_bivariada(renda, 'idade', 'renda')


analise_bivariada(renda, 'estado_civil', 'renda')


analise_bivariada(renda, 'educacao', 'estado_civil')
```

## Etapa 3 Crisp-DM: Preparação dos dados
Nessa etapa realizamos tipicamente as seguintes operações com os dados:

 - **seleção**: Já temos os dados selecionados adequadamente?
 - **limpeza**: Precisaremos identificar e tratar dados faltantes
 - **construção**: construção de novas variáveis
 - **integração**: Temos apenas uma fonte de dados, não é necessário integração
 - **formatação**: Os dados já se encontram em formatos úteis?



<span style="color:red">ESCREVER AQUI</span>


```python


# Exibindo o resumo das colunas para verificar os tipos de dados e valores faltantes
print(renda.info())

# Exibindo uma descrição estatística das variáveis numéricas
print(renda.describe())

# Exibindo a porcentagem de valores faltantes por coluna
missing_data = renda.isnull().mean() * 100
print(missing_data[missing_data > 0].sort_values(ascending=False))

# 1. **Seleção de dados**
# Neste caso, as variáveis já estão selecionadas no dataset "previsao_de_renda.csv".
# Para este projeto, consideraremos as variáveis que influenciam diretamente na previsão de renda.

# 2. **Limpeza dos dados**

# Tratando os valores faltantes nas colunas 'tempo_emprego' (substituindo por mediana)
if renda['tempo_emprego'].isnull().sum() > 0:
    renda['tempo_emprego'].fillna(renda['tempo_emprego'].median(), inplace=True)

# Confirmando que não há mais valores faltantes em 'tempo_emprego'
print(renda['tempo_emprego'].isnull().sum())

# Verificando se existem outros dados faltantes e removendo/ajustando se necessário
# Nesse caso, por exemplo, poderíamos remover linhas com valores faltantes em colunas essenciais
renda.dropna(subset=['renda'], inplace=True)

# 3. **Construção de novas variáveis**
# Exemplo: Criando uma nova variável que indica se o cliente é jovem (idade <= 30)
renda['jovem'] = renda['idade'] <= 30

# Exemplo: Agrupando a variável 'tempo_emprego' em faixas
bins = [0, 2, 5, 10, 20, 30]
labels = ['0-2 anos', '2-5 anos', '5-10 anos', '10-20 anos', '20-30 anos']
renda['faixa_tempo_emprego'] = pd.cut(renda['tempo_emprego'], bins=bins, labels=labels, include_lowest=True)

# 4. **Integração de dados**
# Não é necessária neste projeto, já que temos uma única fonte de dados.

# 5. **Formatação dos dados**
# Verificando se os dados já estão formatados corretamente para análise e modelagem
# Vamos garantir que todas as variáveis categóricas estejam no formato correto
categ_vars = ['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']
for var in categ_vars:
    renda[var] = renda[var].astype('category')

# Exibindo as primeiras linhas do dataset após as modificações
print(renda.head())
```
